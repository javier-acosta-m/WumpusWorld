## Build WumpusWorld on open ai gym environment
The objetive of create an environment of the WUmpusWold in OpenAI Gymasium to be able to use standard libraries for training

```python
# Append the virtual upper folder to the system path
import sys
sys.path.append("../")

# Imports
import logging
import gymnasium as gym
from stable_baselines3.common.env_checker import check_env
import gym_wumpus.envs
from gym_wumpus.envs.wumpus.config import DEFAULT_WUMPUS_WORLD_CONFIG, ConfigField

# Setup the logger
logger = logging.getLogger(gym_wumpus.LOGGER_NAME)
logging.basicConfig(format='%(levelname)s:%(message)s', level=logging.INFO, stream=sys.stdout)
logger.setLevel('INFO')

# Create environment
env_cfg = DEFAULT_WUMPUS_WORLD_CONFIG
env_cfg[ConfigField.PIT_PROB.name] = 0.2
env = gym.make('gym_wumpus/WumpusEnv-v0', render_mode="ansi", config=env_cfg)

# Check the environment interface
check_env(env, warn=True)

# Create a new instance of WUMPUS World
observation, info = env.reset(seed=43)


# Run a sample test
done = False
step = 0
while not done:
    
    # Sample a random action from the list of available actions
    action = env.action_space.sample()

    # Perform this action on the environment
    observation, step_reward, terminated, truncated, info = env.step(action)
    done = terminated or truncated
    logger.info("Step {0} reward {1}".format(step, step_reward))
    step += 1
    
    # print the new state
    env.render()    
    
    
# end this instance of the taxi environment
env.close()
logger.info("Done")
```

## Ancillary Functions
Definition of function to plot results and analyze the runs

```python
import pandas as pd

def plot_with_mov_avg(df, column, window_size):
    column_mov_avg = column + "_MA"  
    df[column].plot()
    df[column_mov_avg] = df[column].rolling(window_size).mean()
    ax = df[column_mov_avg].plot(title= column + " moving average window {0}".format(window_size))
    
```
